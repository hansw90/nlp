# Softmax 함수

### 활성화 함수 

Softmax 함수는 입력 받은 값을 0~1 사이의 출력으로 모두 정규화 하며 출력값들의 총합을 항상 1이 되도록 만들어주는 함수이다.
분류하고 싶은 클래수의 수 만큼 출력으로 구성한다.
가장 큰 출력값을 받은 클래스가 확률이 가장 높은것으로 이용된다.


Softmax = f(x)_t = e^(x_t) / sigma(e^(x_k) 


```python
import numpy as np
import matplotlib.pyplot as plt
 
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
 
x = np.array([1.0,1.0,2.0])
 
y = softmax(x)
 
print(np.sum(y))
 
ratio = y
labels = y
 
plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```
